```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```

```python
osaeg1_df = pd.read_excel("C:/Users/홍정화/Desktop/abc_Camp_BigData/osaeg1.xlsx")
osaeg2_df = pd.read_excel("C:/Users/홍정화/Desktop/abc_Camp_BigData/osaeg2.xlsx")
unam_df = pd.read_excel("C:/Users/홍정화/Desktop/abc_Camp_BigData/unam.xlsx")
osancityhall_df = pd.read_excel("C:/Users/홍정화/Desktop/abc_Camp_BigData/osancityhall.xlsx")
```

```python
def sum_total(dataframe):
    total_day = []
    for i in range(1,7):
        if i==1:
            for j in range(0,31):
                total_day.append(dataframe[str(i)+"월"][j])
        if i==2:
            for j in range(0,28):
                total_day.append(dataframe[str(i)+"월"][j])
        if i==3:
            for j in range(0,31):
                total_day.append(dataframe[str(i)+"월"][j])
        if i==4:
            for j in range(0,30):
                total_day.append(dataframe[str(i)+"월"][j])
        if i==5:
            for j in range(0,31):
                total_day.append(dataframe[str(i)+"월"][j])
        if i==6:
            for j in range(0,30):
                total_day.append(dataframe[str(i)+"월"][j])
    return total_day
```

```python
sum_total_df = pd.DataFrame()
sum_total_df["오색시장 제1공영주차장"]=sum_total(osaeg1_df)
sum_total_df["오색시장 제2공영주차장"]=sum_total(osaeg2_df)
sum_total_df["오산시 부설주차장"]=sum_total(osancityhall_df)
sum_total_df["운암공영주차장"]=sum_total(unam_df)
```
