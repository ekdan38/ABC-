# 각 주차장별 붐비는시간

```python
sum_peak_df = pd.DataFrame()
sum_peak_df["시간(1시간)"]=[]
sum_peak_df["오색시장 제1공영주차장"]=[]
sum_peak_df["오색시장 제2공영주차장"]=[]
sum_peak_df["오산시 부설주차장"]=[]
sum_peak_df["운암공영주차장"]=[]


peak_time=[]
for i in range(0,3):
    for j in range(0,10):
        if i==2 and j>3:
            break
        else:
            a = str(i)+str(j)+":00:00~"+str(i)+str(j)+":59:59"
        peak_time.append(a)
sum_peak_df["시간(1시간)"]=peak_time
```

```python
def parking_sum(parking_df, parking_name):    
    parking = []
    for i in range(0,24):
        n = parking_df["입차수"][i]
        parking.append(n)
    sum_peak_df[parking_name] = parking
```

```python
parking_sum(osaeg1_peak_df, "오색시장 제1공영주차장")
parking_sum(osaeg2_peak_df, "오색시장 제2공영주차장")
parking_sum(unam_peak_df, "운암공영주차장")
parking_sum(osancityhall_peak_df, "오산시 부설주차장")
```

```python
sum_peak_df["오색시장 제1공영주차장"]= round(sum_peak_df["오색시장 제1공영주차장"], 2)
sum_peak_df["오색시장 제2공영주차장"]= round(sum_peak_df["오색시장 제2공영주차장"], 2)
sum_peak_df["오산시 부설주차장"]= round(sum_peak_df["오산시 부설주차장"], 2)
sum_peak_df["운암공영주차장"]= round(sum_peak_df["운암공영주차장"], 2)
```

# 각 주차장별 회전률

```python
rotation_df = pd.DataFrame()

rotation_df["시간(1시간)"]=[]
rotation_df["오색시장 제1공영주차장"]=[]
rotation_df["오색시장 제2공영주차장"]=[]
rotation_df["오산시 부설주차장"]=[]
rotation_df["운암공영주차장"]=[]

peak_time=[]
for i in range(0,3):
    for j in range(0,10):
        if i==2 and j>3:
            break
        else:
            a = str(i)+str(j)+":00:00~"+str(i)+str(j)+":59:59"
        peak_time.append(a)
rotation_df["시간(1시간)"]=peak_time
```

```python
def rotation_sum(parking_df, parking_name):    
    parking = []
    for i in range(0,24):
        n = parking_df["회전률"][i]
        parking.append(n)
    rotation_df[parking_name] = parking
```

```python
rotation_sum(osaeg1_peak_df, "오색시장 제1공영주차장")
rotation_sum(osaeg2_peak_df, "오색시장 제2공영주차장")
rotation_sum(unam_peak_df, "운암공영주차장")
rotation_sum(osancityhall_peak_df, "오산시 부설주차장")
```
