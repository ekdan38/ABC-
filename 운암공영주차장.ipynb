## 이용률 분석하기 위해 입차시각, 출차시각 나누기

```python
df["입차년도"] = df["입차시각"].dt.year.astype('Int64')
df["입차달"] = df["입차시각"].dt.month.astype('Int64')
df["입차일"] = df["입차시각"].dt.day.astype('Int64')
df["입차시"] = df["입차시각"].dt.hour.astype('Int64')
df["입차분"] = df["입차시각"].dt.minute.astype('Int64')
df["입차초"] = df["입차시각"].dt.second.astype('Int64')

df["출차년도"] = df["출차시각"].dt.year.astype('Int64')
df["출차달"] = df["출차시각"].dt.month.astype('Int64')
df["출차일"] = df["출차시각"].dt.day.astype('Int64')
df["출차시"] = df["출차시각"].dt.hour.astype('Int64')
df["출차분"] = df["출차시각"].dt.minute.astype('Int64')
df["출차초"] = df["출차시각"].dt.second.astype('Int64')
```

# 입차시각이 출차시각보다 늦거나 같은 데이터 삭제 
## ex) 입차시각 2022-01-01 00:31:00 출차시각 2022-01-01 00:12:15

```python
for i in range(len(df_2022)):
    In = df_2022["입차시각"][i]
    Out =df_2022["출차시각"][i]
    if(In>=Out):
        df_2022 = df_2022.drop(i)
```

```python
df_2022 = df_2022.reset_index(drop=True)
```

# 이용시간 계산 (출차시각 - 입차시각)

```python
usetime =[]
for i in range(len(df_2022)):
    In = df_2022["입차시각"][i]
    Out = df_2022["출차시각"][i]
    datediff = Out-In
    usetime.append(datediff)
df_2022["이용시간"]=usetime
```

```python
df_2022
```

# 평균 이용시간  => usetime_avg

```python
usetime_sum = df_2022["이용시간"].sum() 
usetime_sum
```

```python
usetime_avg = usetime_sum/len(df_2022["이용시간"])
usetime_avg
```

# 주차이용 효율 계산

```python
real_day = 0
month_entrance = month_entrance_df["월별 입차 수"].sum()
for i in range(1,7):
    real_day += len(all_entrance_df[all_entrance_df[str(i)+"월"] !=0])
```

```python
avg_parking = month_entrance/real_day
avg_parking
```

```python
parking_n = 593 #주차용량
```

```python
usetime_avg = usetime_avg.total_seconds()
```

```python
usetime_avg
```

```python
Utilization_Efficiency = (avg_parking * usetime_avg)/(parking_n * 24)
```

```python
Utilization_Efficiency = round(Utilization_Efficiency/360,2)
Utilization_Efficiency
```

# 붐비는 시간 (1시간 기준) 
## data 수가 너무 많아 실행시간이 너무 오래 걸리기 때문에 각 월에서 1개의 주말 data 추출

```python
peak_df = pd.DataFrame()
peak_df["시간(1시간)"]=[]
peak_df["입차수"]=[]
peak_df["회전률"]=[]

peak_time=[]
for i in range(0,3):
    for j in range(0,10):
        if i==2 and j>3:
            break
        else:
            a = str(i)+str(j)+":00:00~"+str(i)+str(j)+":59:59"
        peak_time.append(a)
peak_df["시간(1시간)"]=peak_time
```

```python
def holiday_df(dataframe, month, day):
    dataframe = df_2022[df_2022["입차달"]==month]
    dataframe = dataframe[dataframe["입차일"]==day]
    dataframe = dataframe.reset_index(drop=True)
    return dataframe
```

```python
jan_holiday_df = pd.DataFrame()
feb_holiday_df = pd.DataFrame()
mar_holiday_df = pd.DataFrame()
apr_holiday_df = pd.DataFrame()
may_holiday_df = pd.DataFrame()
jun_holiday_df = pd.DataFrame()

jan_holiday_df = holiday_df(jan_holiday_df, 1, 29)
feb_holiday_df = holiday_df(feb_holiday_df, 2, 12)
mar_holiday_df = holiday_df(mar_holiday_df, 3, 29)
apr_holiday_df = holiday_df(apr_holiday_df, 4, 23)
may_holiday_df = holiday_df(may_holiday_df, 5, 29)
jun_holiday_df = holiday_df(jun_holiday_df, 6, 18)
```

```python
def time_use(dataframe, month_list):
    for i in range(0,24):
        for j in range(len(dataframe)):
            use = len(dataframe[dataframe["입차시"]==i])
        month_list.append(use)
    return month_list
```

```python
jan = []
feb = []
mar = []
apr = []
may = []
jun = []

jan_use = time_use(jan_holiday_df, jan)
feb_use = time_use(feb_holiday_df, feb)
mar_use = time_use(mar_holiday_df, mar)
apr_use = time_use(apr_holiday_df, apr)
may_use = time_use(may_holiday_df, may)
jun_use = time_use(jun_holiday_df, jun)
```

```python
sum_use = [jan_use[i] + feb_use[i] + mar_use[i] + apr_use[i] + may_use[i] + jun_use[i] for i in range(len(jan_use))]
```

```python
n = 6
sum_use = [x / n for x in sum_use]
```

```python
peak_df["입차수"]=sum_use
```
